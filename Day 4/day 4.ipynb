# Counter using Counter Buttons

```python
from time import sleep
from pynq.overlays.base import BaseOverlay
 
def main():
    base = BaseOverlay('base.bit')
    # LED access
 
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
 
    counter = 0
    for i in range(0,16):
        bit1 = counter&1
        bit2 = (counter>>1)&1
        bit3 = (counter>>2)&1
        bit4 = (counter>>3)&1
 
        if bit1==1:
            base.leds[0].on()
        else:
            base.leds[0].off()
 
 
        if bit2==1:
            base.leds[1].on()
        else:
            base.leds[1].off()
 
 
        if bit3==1:
            base.leds[2].on()
        else:
            base.leds[2].off()
 
 
        if bit4==1:
            base.leds[3].on()
        else:
            base.leds[3].off()
 
        print(counter)
        sleep(1)
        counter+=1
 
 
main()
```

# Counter Continuously Working

```python

from time import sleep
from pynq.overlays.base import BaseOverlay
 
def main():
    base = BaseOverlay('base.bit')
    # LED access
 
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
 
    print("Program started")
    counter = 0
 
    while True:
        switch1 = base.switches[0].read()
        if switch1==1:
            print("Counter Turned on")
            for i in range(0,16):
                bit1 = counter&1
                bit2 = (counter>>1)&1
                bit3 = (counter>>2)&1
                bit4 = (counter>>3)&1
 
                if bit1==1:
                    base.leds[0].on()
                else:
                    base.leds[0].off()
 
 
                if bit2==1:
                    base.leds[1].on()
                else:
                    base.leds[1].off()
 
 
                if bit3==1:
                    base.leds[2].on()
                else:
                    base.leds[2].off()
 
 
                if bit4==1:
                    base.leds[3].on()
                else:
                    base.leds[3].off()
 
                print(counter)
                sleep(1)
                counter+=1
            switch1 = base.switches[0].read()
        else:
            print("Counter Turned off")
main()
```

# Logic Gates

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    #and logic code 
    while True:
        switch1 = base.switches[0].read()
        switch2 = base.switches[1].read()
        
        result1 = switch1 & switch2
        result2 = switch1 or switch2
        result3 = not(switch1 or switch2)
        result4 = not(switch1 & switch2)
        
        print(result4)
        print(result2)
        if result1 == 1:
            base.leds[0].on()
            print ('output = and true')
            
        else:
            base.leds[0].off()
            
        if result2 == 1:
            base.leds[1].on()
            print("output= or true")
            
        else:
            base.leds[1].off()
            
        if result3 == 1:
            base.leds[2].on()
            print("output= exor true")
            
        else:
            base.leds[2].off()

        if result4 == 1:
            base.leds[3].on()
            print("output= not true")
        
        else:
            base.leds[3].off()
            print("output=false (invalid)")

print("output=false (invalid)")
main()
```

# Binary to Gray code

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    print("Program started")
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    #and logic code 
    while True:
        switch2 = base.switches[0].read()  # MSB
        switch1 = base.switches[1].read()  # LSB

        # Binary → Gray conversion
        gray1 = switch1
        gray0 = switch1 ^ switch2

        # Output to LEDs
        if gray1 == 1:
            base.leds[0].on()
            print("Gray MSB = 1")
        else:
            base.leds[0].off()

        if gray0 == 1:
            base.leds[1].on()
            print("Gray LSB = 1")
        else:
            base.leds[1].off()
            
main()
```

# Binary to Excess 3

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    print("Program Started")

    # Initial LED blink
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)

    # Binary to Excess-3 logic
    while True:
        A = base.switches[1].read()
        B = base.switches[0].read()

        # Excess-3 logic
        E0 = int(not B)
        E1 = int(not (A ^ B))   # XNOR
        E2 = A | B
        E3 = 0

        # Display on LEDs
        base.leds[0].write(E0)
        base.leds[1].write(E1)
        base.leds[2].write(E2)
        base.leds[3].write(E3)

        # Print for understanding
        print(f"Input (Binary): {A}{B}")
        print(f"Excess-3 Output: {E3}{E2}{E1}{E0}")
        print("------------------------")

        sleep(0.2)

main()
```

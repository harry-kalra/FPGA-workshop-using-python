# AND Gate

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    print('AND Gate simulation started')
    while True:
        switch1 = base.switches[0].read()
        switch2 = base.switches[1].read()
        
        result = switch1 & switch2
        if result == 1:
            base.leds[0].on()
            print('output and = true')
            
        else:
            base.leds[0].off()
            print('output and = false')
```

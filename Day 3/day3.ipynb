# Half Adder

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    #and logic code 
    while True:
        switch1 = base.switches[1].read()
        switch2 = base.switches[0].read()

        sum = switch1 ^ switch2
        carry = switch1 and switch2
        if sum == 1:
            base.leds[0].on()
            print('sum = 1')
        else:
            base.leds[0].off()

        if carry == 1:
            base.leds[1].on()
            print('carry = 1')
        else:
            base.leds[1].off() 
```

# Half Subtractor

```python
from time import sleep
from pynq.overlays.base import BaseOverlay
 
def main():
    base = BaseOverlay('base.bit')
    # LED access
 
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
 
    #and logic code 
    while True:
        switch1 = base.switches[0].read()
        switch2 = base.switches[1].read()
 
        difference =  switch1^switch2
        borrow = (~(switch1)&1)&switch2
 
        if difference==1:
            base.leds[0].on()
            print ('sum_output = 1')
 
        else:
            base.leds[0].off()
            print("Sum output = 0")
 
        if borrow== 1:
            base.leds[1].on()
            print("carry= 1")
 
        else:
            base.leds[1].off()
            print("carry = 0")
main()
```

# Multiplexer

```python
from pynq.overlays.base import BaseOverlay
import time

# Load overlay
base = BaseOverlay("base.bit")

def run_mux():
    print("4:1 MUX running (Switch-controlled)")

    while True:
        # Inputs
        i0 = base.switches[0].read()
        i1 = base.switches[1].read()
        i2 = base.switches[2].read()
        i3 = base.switches[3].read()

        # Select lines
        s0 = base.switches[4].read()
        s1 = base.switches[5].read()

        sel = (s1 << 1) | s0

        # MUX Logic
        if sel == 0:
            out = i0
        elif sel == 1:
            out = i1
        elif sel == 2:
            out = i2
        elif sel == 3:
            out = i3

        # Output to LED0
        base.leds[0].write(out)

        # Debug print (optional)
        print(f"S1S0={s1}{s0} → Output={out}")

        time.sleep(0.1)
```
